+ Tính lợi nhuận gộp của doanh nghiệp.

### **4) Doanh thu cửa hàng**

image.png

```python
# Input
# b1: số mặt hàng loại một (int)
b1 = int(input("Số mặt hàng một: "))
# c1: giá trên một b1 (int)
c1 = int(input("Giá của một b1: "))
# e1: số mặt hàng loại 1 còn lại (int)
e1 = int(input("Số mặt hàng loại 1 còn lại: "))

# b2: số mặt hàng loại hai (int)
b2 = int(input("Số mặt hàng hai: "))
# c2: giá trên một b2 (int)
c2 = int(input("Giá của một b2: "))
# e2: số mặt hàng loại 2 còn lại (int)
e2 = int(input("Số mặt hàng loại 2 còn lại: "))

# Process:
# Doanh thư thực tế
# Mặt hàng 1
thuc_te_1 = (b1 - e1) * c1
toi_da_1 = b1 * c1
# Mặt hàng 2:
thuc_te_2 = (b2 - e2) * c2
toi_da_2 = b2 * c2

thuc_te = thuc_te_1 + thuc_te_2
toi_da = toi_da_1 + toi_da_2

# Output:
# 1. Doanh thu thực tế
print(f"Doanh thu thực tế: {thuc_te}")
# 2. Doanh thu tối đa có thể đạt được
print(f"Doanh thu tối đa: {toi_da}")
```

Đề bài: Nhập vào một số tiền bằng USD và tỉ giá hiện tại. Hãy chuyển đổi số tiền đó sang VNĐ.

### **4) Chuyển đổi đơn vị**

image.png

```python
# Input: Số knut => integer => int
knut = int(input("Nhập số knut: "))

# Process: đổi từ knut sang galleon và sickle
# Đổi sang galleon: chia knut cho 493 lấy nguyên
galleon = knut // (17*29)
sickle = (knut % (17*29)) // 29
remainder_knuts = (knut % (17*29)) % 29

# Output: ? galleon ? sickle ? knut
print(f"{galleon} galleons, {sickle} sickles, {remainder_knuts} knuts")
```
