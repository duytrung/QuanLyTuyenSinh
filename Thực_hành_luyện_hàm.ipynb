**Ph√¢n t√≠ch bi·∫øn ƒë·ªông chi ph√≠ nguy√™n v·∫≠t li·ªáu tr·ª±c ti·∫øp**

‚Äì X√°c ƒë·ªãnh ch·ªâ ti√™u ph√¢n t√≠ch

+ C0 = Q1*m0*G0
+ C1 = Q1*m1*G1
+ C0: Chi ph√≠ nguy√™n v·∫≠t li·ªáu tr·ª±c ti·∫øp ƒë·ªãnh m·ª©c
+ C1: Chi ph√≠ nguy√™n v·∫≠t li·ªáu tr·ª±c ti·∫øp th·ª±c t·∫ø
+ Q1: S·ªë l∆∞·ª£ng s·∫£n ph·∫©m s·∫£n xu·∫•t th·ª±c t·∫ø
+ m0: L∆∞·ª£ng nguy√™n v·∫≠t li·ªáu tr·ª±c ti·∫øp ƒë·ªãnh m·ª©c s·∫£n xu·∫•t 1 s·∫£n ph·∫©m

+ G0: Gi√° mua ƒë·ªãnh m·ª©c 1 ƒë∆°n v·ªã nguy√™n v·∫≠t li·ªáu tr·ª±c ti·∫øp
+ G1: Gi√° mua th·ª±c t·∫ø 1 ƒë∆°n v·ªã nguy√™n v·∫≠t li·ªáu tr·ª±c ti·∫øp

‚Äì X√°c ƒë·ªãnh ƒë·ªëi t∆∞·ª£ng ph√¢n t√≠ch

‚Äì Bi·∫øn ƒë·ªông chi ph√≠ (‚àÜC):
+ ‚àÜC = C1 ‚Äì C0
+ ‚àÜC > 0: b·∫•t l·ª£i
+ ‚àÜC <= 0: thu·∫≠n l·ª£i

‚Äì X√°c ƒë·ªãnh ·∫£nh h∆∞·ªüng c·ªßa c√°c nh√¢n t·ªë (c√¥ng th·ª©c k·∫ø to√°n qu·∫£n tr·ªã)

L∆∞·ª£ng nguy√™n v·∫≠t li·ªáu tr·ª±c ti·∫øp ti√™u hao ‚Äì bi·∫øn ƒë·ªông l∆∞·ª£ng (‚àÜCm):

+ ‚àÜCm <= 0: thu·∫≠n l·ª£i

Gi√° mua nguy√™n v·∫≠t li·ªáu tr·ª±c ti·∫øp ‚Äì bi·∫øn ƒë·ªông gi√° (‚àÜCG):
